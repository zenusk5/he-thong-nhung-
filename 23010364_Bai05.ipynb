```
#bai2
weights_input_hidden = [
    [0.1, 0.1, 0.1],
    [0.2, 0.2, 0.2],
    [0.1, 0.2, 0.1]
]
weights_hidden_output = [0.1, 0.1, 0.1]

def dot_product(vec1, vec2):
    return sum(v1 * v2 for v1, v2 in zip(vec1, vec2))

def hidden_layer_forward(inputs, w_in_hid):
    return [dot_product(inputs, neuron_w) for neuron_w in w_in_hid]

def output_layer_forward(hidden_outputs, w_hid_out):
    return dot_product(hidden_outputs, w_hid_out)

def neural_network(inputs, w_in_hid, w_hid_out):
    hidden_outputs = hidden_layer_forward(inputs, w_in_hid)
    prediction = output_layer_forward(hidden_outputs, w_hid_out)
    return hidden_outputs, prediction

def compute_mean_loss_2layer(inputs_list, targets, w_in_hid, w_hid_out):
    total_loss = 0
    for inputs, target in zip(inputs_list, targets):
        _, pred = neural_network(inputs, w_in_hid, w_hid_out)
        total_loss += (pred - target) ** 2
    return total_loss / len(inputs_list)

def train_one_sample_linear(inputs, truth, w_in_hid, w_hid_out, alpha=0.01):
    hidden_outputs, prediction = neural_network(inputs, w_in_hid, w_hid_out)
    error = prediction - truth

    w_hid_out_gradients = [error * h for h in hidden_outputs]

    w_in_hid_gradients = []
    for j in range(len(w_in_hid)):
        hidden_error = error * w_hid_out[j]
        neuron_gradients = [hidden_error * i for i in inputs]
        w_in_hid_gradients.append(neuron_gradients)
    for j in range(len(w_hid_out)):
        w_hid_out[j] -= alpha * w_hid_out_gradients[j]

    for j in range(len(w_in_hid)):
        for k in range(len(inputs)):
            w_in_hid[j][k] -= alpha * w_in_hid_gradients[j][k]

    return w_in_hid, w_hid_out

epochs = 30
alpha = 0.01
for epoch in range(epochs):
    for inputs, truth in zip(dataset_inputs, true_scores):
        weights_input_hidden, weights_hidden_output = train_one_sample_linear(
            inputs, truth, weights_input_hidden, weights_hidden_output, alpha
        )

    # Tính toán và in ra lỗi trung bình sau mỗi epoch
    loss = compute_mean_loss_2layer(dataset_inputs, true_scores, weights_input_hidden, weights_hidden_output)
    print(f"Epoch {epoch:2d}: Mean Loss = {loss:.4f}")
```

```

def relu(x):
    return x if x > 0 else 0

def relu_derivative(x):
    return 1 if x > 0 else 0

def dot_product(vec1, vec2):
    return sum(v1 * v2 for v1, v2 in zip(vec1, vec2))

def hidden_layer_forward_with_activation(inputs, weights_input_hidden):
    hidden_outputs_activated = []
    for neuron_weights in weights_input_hidden:

        pre_act = dot_product(inputs, neuron_weights)

        hidden_outputs_activated.append(relu(pre_act))
    return hidden_outputs_activated


def neural_network(inputs, w_in_hid, w_hid_out):
    hidden_outputs = hidden_layer_forward_with_activation(inputs, w_in_hid)
    prediction = dot_product(hidden_outputs, w_hid_out)
    return hidden_outputs, prediction

def train_one_sample_with_activation(inputs, true_score, w_in_hid, w_hid_out, alpha=0.01):

    hidden_pre_activation = [dot_product(inputs, neuron_w) for neuron_w in w_in_hid]
    hidden_outputs = [relu(x) for x in hidden_pre_activation]
    prediction = dot_product(hidden_outputs, w_hid_out)


    error = prediction - true_score


    w_hid_out_gradients = [error * h for h in hidden_outputs]


    w_in_hid_gradients = []
    for j in range(len(w_in_hid)):
        hidden_error = error * w_hid_out[j]

        hidden_delta = hidden_error * relu_derivative(hidden_pre_activation[j])


        neuron_gradients = [hidden_delta * incoming_input for incoming_input in inputs]
        w_in_hid_gradients.append(neuron_gradients)

    next_w_hid_out = [w_hid_out[j] - alpha * w_hid_out_gradients[j] for j in range(len(w_hid_out))]

    next_w_in_hid = []
    for j in range(len(w_in_hid)):
        updated_row = [w_in_hid[j][k] - alpha * w_in_hid_gradients[j][k] for k in range(len(inputs))]
        next_w_in_hid.append(updated_row)

    sample_loss = error ** 2
    return next_w_in_hid, next_w_hid_out, sample_loss
```

```


def neural_network_linear(inputs, w_in_hid, w_hid_out):
    hidden_outputs = [dot_product(inputs, neuron_w) for neuron_w in w_in_hid]
    prediction = dot_product(hidden_outputs, w_hid_out)
    return hidden_outputs, prediction

def compute_mean_loss_linear(inputs_list, targets, w_in_hid, w_hid_out):
    total_loss = 0
    for inputs, target in zip(inputs_list, targets):
        _, pred = neural_network_linear(inputs, w_in_hid, w_hid_out)
        total_loss += (pred - target) ** 2
    return total_loss / len(inputs_list)

def train_one_sample_linear(inputs, truth, w_in_hid, w_hid_out, alpha=0.01):
    hidden_outputs, prediction = neural_network_linear(inputs, w_in_hid, w_hid_out)
    error = prediction - truth

    w_hid_out_gradients = [error * h for h in hidden_outputs]
    w_in_hid_gradients = []
    for j in range(len(w_in_hid)):
        hidden_error = error * w_hid_out[j]
        neuron_gradients = [hidden_error * i for i in inputs]
        w_in_hid_gradients.append(neuron_gradients)

    next_w_hid_out = [w_hid_out[j] - alpha * w_hid_out_gradients[j] for j in range(len(w_hid_out))]
    next_w_in_hid = [[w_in_hid[j][k] - alpha * w_in_hid_gradients[j][k] for k in range(len(inputs))] for j in range(len(w_in_hid))]
    return next_w_in_hid, next_w_hid_out


def compute_mean_loss_relu(inputs_list, targets, w_in_hid, w_hid_out):
    total_loss = 0
    for inputs, target in zip(inputs_list, targets):
        _, pred = neural_network(inputs, w_in_hid, w_hid_out)
        total_loss += (pred - target) ** 2
    return total_loss / len(inputs_list)

dataset_inputs = [[1, 5, 5], [2, 6, 6], [3, 7, 6], [4, 6, 7], [5, 7, 8], [6, 8, 9]]
true_scores = [3, 5, 6, 7, 9, 11]
```
