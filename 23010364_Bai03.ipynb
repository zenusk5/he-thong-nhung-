```
weights = [0.1, 0.1, 0.1]

for i in range(len(dataset_inputs)):
    inputs = dataset_inputs[i]
    truth = truth_scores[i]

    prediction = neural_network(inputs, weights)

    error = prediction - truth

    print(i, prediction, truth, error)
```

```
def squared_error_from_pred(prediction, truth):
  return (prediction - truth) ** 2
```

```
total_loss = 0
for i in range(len(dataset_inputs)):
  pre = neural_network(dataset_inputs[i], weights)
  error_sq = squared_error_from_pred(pre, truth_scores[i])
  total_loss = total_loss + error_sq
  print (total_loss)
```

```
mean_loss = total_loss / len(dataset_inputs)
print (mean_loss)
```

```
def computer_mean_loss(dataset_inputs , truth_scores , weights):
  total_loss = 0
  for i in range(len(dataset_inputs)):
    pre = neural_network(dataset_inputs[i] , weights)
    total_loss = total_loss + squared_error_from_pred(pre , truth_scores[i])
  return total_loss / len(dataset_inputs)
```

```
weights_to_try = [
    [0.0, 0.0, 0.0],
    [0.1, 0.1, 0.1],
    [0.2, 0.1, 0.1],
    [0.5, 0.2, 0.1],
    [1.0, 0.2, 0.1],
    [0.5, 0.5, 0.5]
]
for w in weights_to_try:
    loss = computer_mean_loss(dataset_inputs, truth_scores, w)
    print (loss)
```

```
weights = [0.1, 0.1, 0.1]
alpha = 0.01

inputs = dataset_inputs[0]
truth = truth_scores[0]

prediction = neural_network(inputs , weights)
error = prediction - truth

for j in range (len(weights)):
  weights[j] = weights[j] - alpha * error * inputs[j]

print (weights)
```

```

def update_weights_one_sample(weights, inputs, truth, alpha):
    prediction = neural_network(inputs, weights)
    error = prediction - truth

  #tao ban sao
    for j in range(len(weights)):
        weights[j] = weights[j] - (alpha * error * inputs[j])
    return weights
```

```

weights = [0.1, 0.1, 0.1]
alpha = 0.01

# Duyệt qua toàn bộ dữ liệu (1 epoch)
for i in range(len(dataset_inputs)):
    weights = update_weights_one_sample(weights, dataset_inputs[i], truth_scores[i], alpha)

print("Weights sau 1 epoch:", weights)
```

```
weights = [0.1, 0.1, 0.1]
alpha = 0.01

for epoch in range(30):
  for i in range(len(dataset_inputs)):
    weights = update_weights_one_sample(weights, dataset_inputs[i], truth_scores[i], alpha)
    loss = computer_mean_loss(dataset_inputs, truth_scores, weights)
    print(epoch, i, loss)
```
