```
#bai12
def squared_error(error):
  return error * error
```

```
#bai13
def error_squared(prediction , truth):
  return (prediction - truth) ** 2
```

```
#bai14
weight = 0.5
for i in range(len(study_hours)):
  input = study_hours[i]
  truth = true_scores[i]

  prediction = neural_network(input, weight)
  squared_error = error_squared(prediction , truth)

  print (input , truth , prediction , squared_error)
```

```
#bai 15 va bai16
weight = 0.5
total_loss = 0
for i in range(len(study_hours)):
  prediction = neural_network(study_hours[i], weight)
  total_loss = total_loss + error_squared(prediction , true_scores[i])
mean_loss = total_loss / len(study_hours)
print (mean_loss)
print(total_loss)
```

```
#bai17
def compute_total_loss(inputs , truths , weight):
  total_loss = 0
  for i in range (len(inputs)):
    prediction = neural_network(inputs[i] , weight)
    total_loss = total_loss + error_squared(prediction , truths[i])
  return total_loss



```

```
#bai18
def compute_mean_loss (inputs , truths , weight):
  t_loss = compute_total_loss(inputs, truths , weight)
  return t_loss / len(inputs)
```

```
# Bài 19
weights_to_try = [0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3]

for w in weights_to_try:
    loss = compute_mean_loss(study_hours, true_scores, w)
    print(f"{w:<10} {loss:<10.4f}")
```

```
#bai20
best_loss = None
best_weight = None

for w in weights_to_try:
  loss = compute_mean_loss(study_hours , true_scores,w)
  if best_loss is None or loss < best_loss:
    best_loss = loss
    best_weight = w
print (best_loss , best_weight)
```

```
#bai 21
w1, l1 = None, float('inf')
w2, l2 = None, float('inf')
w3, l3 = None, float('inf')

weights_to_try = [0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3]

for w in weights_to_try:
    loss = compute_mean_loss(study_hours, true_scores, w)


    if loss < l1:

        w3, l3 = w2, l2
        w2, l2 = w1, l1
        w1, l1 = w, loss
    elif loss < l2:

        w3, l3 = w2, l2
        w2, l2 = w, loss
    elif loss < l3:

        w3, l3 = w, loss

print(f"1: Weight = {w1}, Loss = {l1:.4f}")
print(f"2: Weight = {w2}, Loss = {l2:.4f}")
print(f"3: Weight = {w3}, Loss = {l3:.4f}")

best_weight = w1
```

```
sleep_hours = [2, 3, 4, 5, 1, 3.5]
alertness_scores = [4, 6, 8, 10, 2, 7]

weight = 0.1
alpha = 0.01
epochs = 50

for epoch in range(epochs):
    for i in range(len(sleep_hours)):
        input_val = sleep_hours[i]
        truth = alertness_scores[i]

        prediction = input_val * weight
        error = prediction - truth


        weight = weight - (alpha * error * input_val)

print("Weight mới:", weight)

# test
test_sleep = 6
prediction = test_sleep * weight
print("prediction:", prediction)
```
