```
#bai3
weights_input_hidden = [
    [0.1 , 0.1 , 0.1],
    [0.2 , 0.2 , 0.2]
]

print (weights_input_hidden)
print (len(weights_input_hidden))
print (len(weights_input_hidden[0]))
```

```
#bai4
weights_hidden_output = [0.1 , 0.1]
```

```
#bai5
def dot_product (inputs , weights):
  result = 0
  for i in range (len(inputs)):
   result = result + inputs[i] * weights[i]
  return result
```

```
#bai6
sample = [ 1 ,6,7]
hidden_0 = dot_product(sample , weights_input_hidden[0])
print (hidden_0)
```

```
#bai7
def hidden_layer_forward (inputs , weights_input_hidden):
  hidden_outputs = []
  for i in range (len(weights_input_hidden)):
    out = dot_product(inputs , weights_input_hidden[i])
    hidden_outputs.append(out)
  return hidden_outputs

```

```
#bai8
hidden_outputs = hidden_layer_forward(sample , weights_input_hidden)
print (hidden_outputs)
```

Phan C

```
#bai9
def output_layer_forward (hidden_outputs, weights_hidden_output):
  prediction  = dot_product(hidden_outputs , weights_hidden_output)
  return prediction
```

```
from os import pread
#bai10
def neural_network(inputs , w_in_hid , w_hid_out):
  hid_out = hidden_layer_forward(inputs , w_in_hid)
  prediction = output_layer_forward(hid_out , w_hid_out)
  return hid_out , prediction
```

```
for i in range (len(dataset_inputs)):
  sample = dataset_inputs[i]

pred = neural_network(sample , weights_input_hidden , weights_hidden_output)[1]
error = pred - true_scores[i]
print (error)
```

```
#bai16
inputs = dataset_inputs[0]
for h in range(len(weights_input_hidden)):
    for i in range(len(inputs)):
        weights_input_hidden[h][i] -= alpha * hidden_errors[h] * inputs[i]
```

```
#bai17
def train_one_sample (inputs , truth , w_in_hid , w_hid_out , alpha):
  hid_out , prediction = neural_network(inputs , w_in_hid , w_hid_out)
  error = prediction - truth

  hidden_errors = []
  for j in range(len(w_hid_out)):
    hidden_errors.append(error * w_hid_out[j])

  for j in range (len(w_hid_out)):
    w_hid_out[j] = w_hid_out[j] - alpha * error * hid_out[j]

  for h in range(len(w_in_hid)):
    for i in range(len(inputs)):
        w_in_hid[h][i] =w_in_hid[h][i] - alpha * hidden_errors[h] * inputs[i]
  return error ** 2
```

Phan E

```
#bai18
def train_one_epochs(dataset, truths, w_in_hid, w_hid_out, alpha):
    for i in range(len(dataset)) :
        train_one_sample(dataset[i], truths[i], w_in_hid, w_hid_out, alpha)
```

```
#bai19
def compute_mean_loss_2layer(dataset, truths, w_in_hid, w_hid_out):

    total_squared_error = 0

    for i in range(len(dataset)):

        prediction = neural_network(dataset[i],w_in_hid,w_hid_out)[1]

        error = prediction - truths[i]

        squared_error = error ** 2

        total_squared_error += squared_error

    mean_loss = total_squared_error / len(dataset)

    return mean_loss
```

```
alpha = 0.0001
epochs = 30

weights_input_hidden = [
    [0.1, 0.1, 0.1],
    [0.2, 0.2, 0.2]
]

weights_hidden_output = [0.1, 0.1]
for epoch in range(epochs):

    train_one_epochs(
        dataset_inputs,
        true_scores,
        weights_input_hidden,
        weights_hidden_output,
        alpha
    )

    current_loss = compute_mean_loss_2layer(
        dataset_inputs,
        true_scores,
        weights_input_hidden,
        weights_hidden_output
    )

    print(f"Epoch {epoch:<5} | Loss: {current_loss:.6f}")
```

```
#bai21
for i in range(len(dataset_inputs)):
    pred = neural_network(dataset_inputs[i], weights_input_hidden, weights_hidden_output)[1]
    print(f"Mẫu {i}: Thực tế {true_scores[i]} - Dự đoán {pred:.2f}")
```

```
#
for test_alpha in [0.1, 0.001]:

    w_in_hid_test = [
        [0.1, 0.1, 0.1],
        [0.2, 0.2, 0.2]
    ]

    w_hid_out_test = [0.1, 0.1]

    try:

        for epoch in range(11):

            train_one_epoch(
                dataset_inputs,
                true_scores,
                w_in_hid_test,
                w_hid_out_test,
                test_alpha
            )

            l = compute_mean_loss_2layer(
                dataset_inputs,
                true_scores,
                w_in_hid_test,
                w_hid_out_test
            )

            print(f"Epoch {epoch}: Loss = {l:.6f}")

    except OverflowError:

        print("Lỗi OverflowError: Giá trị quá lớn!")
```

```
#bai23
new_w_in_hid = [
    [0.15, 0.15, 0.15],
    [0.25, 0.25, 0.25],
    [0.35, 0.35, 0.35]
]

new_w_hid_out = [0.1, 0.2, 0.3]

print("Đã khởi tạo mạng mới với 3 neuron ẩn.")
print("Weights Input-Hidden:", new_w_in_hid)
print("Weights Hidden-Output:", new_w_hid_out)
```

```
# Cài đặt thông số cho mạng mới
new_alpha = 0.01
new_epochs = 30

print(f"{'Epoch':<10} | {'Mean Loss':<20}")
print("-" * 35)

# Vòng lặp huấn luyện chính xác 30 lần [cite: 150]
for epoch in range(new_epochs):

    # 1. Huấn luyện toàn bộ dataset trong epoch hiện tại [cite: 143-146]
    # Duyệt qua từng mẫu và gọi hàm train_one_sample đã viết ở Bài 17
    for i in range(len(dataset_inputs)):
        train_one_sample(
            dataset_inputs[i],
            true_scores[i],
            new_w_in_hid,
            new_w_hid_out,
            new_alpha
        )

    # 2. Tính toán Mean Squared Error sau khi kết thúc epoch [cite: 147-148]
    loss = compute_mean_loss_2layer(
        dataset_inputs,
        true_scores,
        new_w_in_hid,
        new_w_hid_out
    )

    # 3. In kết quả Mean Loss cho từng epoch (không bỏ sót epoch nào) [cite: 154, 176]
    print(f"Epoch {epoch:<5} | Loss: {loss:.6f}")

# Kiểm tra độ chính xác cuối cùng của mạng 3 neuron ẩn
print("\n--- KIỂM TRA DỰ ĐOÁN CUỐI CÙNG ---")
for i in range(len(dataset_inputs)):
    _, final_pred = neural_network(dataset_inputs[i], new_w_in_hid, new_w_hid_out)
    print(f"Mẫu {i}: Thực tế {true_scores[i]} - Dự đoán: {final_pred:.2f}")
```
